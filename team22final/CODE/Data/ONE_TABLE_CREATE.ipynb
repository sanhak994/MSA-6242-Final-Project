```python
import sqlite3
from sqlite3 import Error
import pandas as pd
database = "Data_1_Table.db"
```

```python
def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
 
    return conn
```

```python
conn = create_connection(database)
```

```python
file = r'Prediction_inflation_adjusted.csv'
clean_data = pd.read_csv(file, encoding='utf-8')
columns = clean_data.columns
c = conn.cursor()
```

```python
clean_data.to_sql('ALL_DATA', conn, if_exists='append', index = False) # Insert the values from the csv file into the table 
```

```python
df = pd.read_sql_query('SELECT * FROM ALL_DATA;', conn)
df.head()
```
